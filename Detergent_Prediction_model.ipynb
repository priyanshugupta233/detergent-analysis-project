```python
import pandas as pd

# Create dataset
data = {
    "Detergent": ["Surf Excel", "Ariel", "Tide", "Nirma", "Ariel", "Tide"],
    "Strain" : ["Oil", "Mud", "Ink", "Mud", "Oil", "Ink"],
    "Amount" : [50, 67, 87,69, 47,98],
    "Time" : [10, 5, 75, 9, 40,78],
    "Score" : [7, 5, 8, 6, 2, 7]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.head()
df.info()
df.describe()
```

```python
print(df.isnull().sum())
```

```python
df.groupby("Detergent")["Score"].mean()
```

```python
import matplotlib.pyplot as plt

df.groupby("Detergent")["Score"].mean().plot(kind= 'bar')
plt.title("Detergent vs Cleaning Score")
plt.show()
```

```python
df.groupby("Strain")["Score"].mean()
```

```python
from sklearn.linear_model import LinearRegression

# Features
X = df[["Amount", "Time"]]

#Target
y = df["Score"]

model = LinearRegression()
model.fit(X, y)

# Predict
print(model.predict([[50, 10]]))
```
